```python
import requests
from bs4 import BeautifulSoup

def scrape_wikipedia_page(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extract the title of the article
        title = soup.find('title').text
        print(f"Title: {title}")
        
        # Extract and print the main content of the article
        content_div = soup.find('div', {'class': 'mw-parser-output'})
        paragraphs = content_div.find_all('p')
        for paragraph in paragraphs:
            print(paragraph.get_text())
        
        # Find and print links to other pages (pagination-like behavior)
        for link in soup.find_all('a', href=True):
            href = link['href']
            if href.startswith('/wiki/') and not href.startswith('/wiki/Category:'):
                print(f"Link: https://en.wikipedia.org{href}")
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

# URL of the Wikipedia page to scrape
wikipedia_url = "https://en.wikipedia.org/wiki/Three_Arrows_Capital"
scrape_wikipedia_page(wikipedia_url)
```
