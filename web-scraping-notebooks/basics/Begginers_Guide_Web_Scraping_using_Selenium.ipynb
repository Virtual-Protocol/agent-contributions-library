USING DEF FUNCTION, CREATE AN INSTANCE OF CHROMEOPTIONS TO CUSTOMIZE BROWSER SETTINGS

```python
def web_driver():
    # Create an instance of ChromeOptions to customize the browser settings
    options = webdriver.ChromeOptions()

    # Enable verbose logging to get detailed log output from the browser
    options.add_argument('--verbose')

    # Run Chrome in headless mode (no graphical user interface)
    options.add_argument('--headless')

    # Disable the sandbox for Chrome; necessary in some environments (e.g., Docker)
    options.add_argument('--no-sandbox')

    # Disable GPU hardware acceleration; useful when running in environments without GPU support
    options.add_argument('--disable-gpu')

    # Set the window size for the headless browser; ensures consistent rendering of pages
    options.add_argument('--window-size=1920,1200')

    # Disable the use of /dev/shm for temporary files; helps in environments with limited shared memory
    options.add_argument('--disable-dev-shm-usage')

    # Initialize the Chrome WebDriver with the specified options
    driver = webdriver.Chrome(options=options)

    # Return the WebDriver instance for use in your tests or automation tasks
    return driver
```

 CALL THE WEB_DRIVER FUNCTION TO CREATE A WEBDRIVER INSTANCE AND ASSIGN IT TO 'DRIVER'

```python
driver = web_driver()
```

NAVIGATE TO THE SPECIFIED URL

```python
driver.get('https://books.toscrape.com/')
```

FIND ALL ELEMENTS THAT MATCH THE SPECIFIED XPATH EXPRESSION

```python
containers = driver.find_elements(By.XPATH, '//article[@class="product_pod"]')
```

ITERATE THROUGH EACH CONTAINER TO EXTRACT INFORMATION

```python
for container in containers:
  title = container.find_element(By.XPATH, './/h3/a').text
  price = container.find_element(By.XPATH, './/p[@class="price_color"]').text
  print(price)
```
